## Histogram distribution of the core shapes from the market
Visible is that most common are the toroid, E, EQ, RM

```python
l_cores = PyMKF.get_core_shapes()
print("Nr. of cores available: " + str(len(l_cores)))

# l_cores is a list of dictionaries, each dictionary contains the following keys:
all_keys = l_cores[0].keys()
df_cores = pd.DataFrame(columns=all_keys)

for idx, item in enumerate(l_cores):
    df_row = pd.DataFrame([item])
    df_cores = pd.concat([df_cores, df_row], ignore_index=True)

# plot the cores histogram based on family column values in df_cores
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df_cores['family'].value_counts().index, df_cores['family'].value_counts().values)
ax.set_xlabel('Core Family')
ax.set_ylabel('Number of Cores')
ax.set_title('Histogram of Core Families')
ax.xaxis.set_tick_params(rotation=45)
ax.grid(True)
plt.show()
```
